```python
from google.cloud import bigquery

client = bigquery.Client(project='xlnetpred')


sql= '''
SELECT 
    size, timestamp, transaction_count, timestamp_month
FROM 
    `bigquery-public-data.crypto_bitcoin.blocks` t

WHERE 
    
    timestamp_month >= '2019-02-01' and timestamp_month <= '2021-05-21'

   
'''
```

```python
df = client.query(sql).to_dataframe()
```

```python
df = df.sort_values('timestamp_month').reset_index(drop=True)
```

```python
df.to_csv('/content/drive/My Drive/Magisterka/dane_final/blocks_data_2021.csv', index = False)
```
