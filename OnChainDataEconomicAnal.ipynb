```python
blocks = pd.read_csv('/content/drive/My Drive/Magisterka/dane_final/blocks_data_2021.csv')
```

```python
blocks['timestamp']=pd.to_datetime(blocks.timestamp, utc = True)
blocks['timestamp'] = blocks['timestamp'].apply(lambda x: x.replace(second=0, microsecond=0, minute=0))
```

```python
blocks.sort_values(by = 'timestamp')
```



*   liczba blokow w godzine
*   sredni size bloku
*   liczba transakcji
*   srednia liczba transakcji w bloku



```python
data1 = blocks.groupby(by = 'timestamp').count().reset_index()[['timestamp', 'size']]
data1.columns = ['timestamp', 'no_blocks']
```

```python
data2 = blocks.groupby(by = 'timestamp').mean().reset_index()[['timestamp', 'size', 'transaction_count']]
data2.columns = ['timestamp', 'mean_size', 'mean_transaction_count']
```

```python
data3 = blocks.groupby(by = 'timestamp').sum().reset_index()[['timestamp','transaction_count']]
data3.columns = ['timestamp', 'sum_transaction']
```

```python
chain_data = data1.merge(data2, how = 'left', left_on = 'timestamp', right_on = 'timestamp').merge(data3, how = 'left', right_on = 'timestamp', left_on = 'timestamp')
```

```python
import matplotlib.pyplot as plt
plt.plot(chain_data.no_blocks)
```

```python
plt.plot(chain_data.mean_size)
```

```python
plt.plot(chain_data.mean_transaction_count)
```

```python
def to_time(i):
  return ':00'.join(('T'.join((eurousd.Date[i], eurousd.Time[i])),'Z'))

to_time_vec = np.vectorize(to_time)

```
